```python
# Assuming you have a DataFrame named 'reviews_df' with a column 'Review' containing review text
df['Characters Count'] = df['Review'].str.len()

# Display the DataFrame with the added 'Characters Count' column
df
```

```python
# Assuming you have a DataFrame named 'df' with a column 'Review' containing text data
def average_word_length(text):
    words = text.split()  # Split the text into words
    if len(words) != 0:
        return sum(len(word) for word in words) / len(words)  # Calculate average word length
    else:
        return None

df['Average Word Length'] = df['Review'].apply(average_word_length)

# Display the DataFrame with the added 'Average Word Length' column
df
```

```python
# Assuming you have a DataFrame named 'df' with a column 'Review' containing text data
def count_special_characters(text):
    special_char_count = 0
    for char in text:
        if not (char.isalpha() or char.isdigit()):
            special_char_count += 1
    return special_char_count

df['Num of Special Characters'] = df['Review'].apply(count_special_characters)

# Display the DataFrame with the added 'Num of Special Characters' column
df
```

Question 4 (10 points): Write python code to rank the text based on text similarity. Based on the text data you used for question 2, design a query to match the most relevant docments. Please use the BERT model to represent both your query and the text data, then calculate the cosine similarity between the query and each text in your data. Rank the similary with descending order.

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from transformers import AutoTokenizer, AutoModel
```

```python
# Load the pre-trained BERT model and tokenizer
model_name = "bert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)
```

```python
# Assuming you have a DataFrame named 'df' with a column 'Review' containing text data
reviews = df['Review'].tolist()
query = "I want to find a good restaurant."

```

```python
# Tokenize the documents and query
tokenized_reviews = [tokenizer(review, padding=True, truncation=True, return_tensors="pt") for review in reviews]
tokenized_query = tokenizer(query, padding=True, truncation=True, return_tensors="pt")
```

```python
# Encode the documents and query using the BERT model
encoded_reviews = [model(**tokens).pooler_output.squeeze().detach().numpy() for tokens in tokenized_reviews]
encoded_query = model(**tokenized_query).pooler_output.squeeze().detach().numpy()

# Calculate cosine similarity between the query and each document
similarities = cosine_similarity([encoded_query], encoded_reviews)

# Create a DataFrame to store the results
results_df = pd.DataFrame({'Review': reviews, 'Similarity': similarities[0]})

# Sort the results by similarity in descending order
sorted_results_df = results_df.sort_values(by='Similarity', ascending=False)

# Display the sorted results
print(sorted_results_df)
```
