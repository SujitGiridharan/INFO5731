Question 3 (10 points): Write python code to collect 1000 articles from Google Scholar (https://scholar.google.com/), Microsoft Academic (https://academic.microsoft.com/home), or CiteSeerX (https://citeseerx.ist.psu.edu/index), or Semantic Scholar (https://www.semanticscholar.org/), or ACM Digital Libraries (https://dl.acm.org/) with the keyword "information retrieval". The articles should be published in the last 10 years (2013-2023).

The following information of the article needs to be collected:

(1) Title

(2) Venue/journal/conference being published

(3) Year

(4) Authors

(5) Abstract

```python
# You code here (Please add comments in the code):

# You code here (Please add comments in the code):
import requests
from bs4 import BeautifulSoup

# Function to scrape and display academic articles from a given URL
def fetch_and_display_academic_articles(search_url):
    # Send a GET request to the URL and receive the response
    http_response = requests.get(search_url)

    # Parse the HTML content of the page using BeautifulSoup
    parsed_html = BeautifulSoup(http_response.text, 'html.parser')

    # Find all the article information sections on the page
    article_sections = parsed_html.find_all('div', class_='gs_ri')

    # Loop through each article to extract and display relevant information
    for article_section in article_sections:
        # Extracting the title of the article
        article_title = article_section.find('h3', class_='gs_rt').text

        # Extracting the venue, year, and authors
        author_venue_year_info = article_section.find('div', class_='gs_a').text.split(' - ')
        publication_year = author_venue_year_info[0]
        author_names = author_venue_year_info[1]
        publication_venue = author_venue_year_info[-1]

        # Extracting the abstract
        article_abstract = article_section.find('div', class_='gs_rs').text

        # Displaying the extracted information
        print(f"Title: {article_title}")
        print(f"Venue: {publication_venue}")
        print(f"Year: {publication_year}")
        print(f"Authors: {author_names}")
        print(f"Abstract: {article_abstract}")
        print("\n---\n")

# Number of articles per search page (usually 10 for Google Scholar)
articles_per_search_page = 10

# Total number of articles to fetch
total_number_of_articles = 1000

# Loop through multiple pages to scrape 1,000 articles
for start_index in range(0, total_number_of_articles, articles_per_search_page):
    # Create the URL for each page of search results
    page_url = f"https://scholar.google.com/scholar?hl=en&as_sdt=0%2C5&q=information+retrieval&as_ylo=2013&as_yhi=2023&start={start_index}"

    # Fetch and display articles from this page
    fetch_and_display_academic_articles(page_url)




```
